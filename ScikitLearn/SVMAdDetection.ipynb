```python
import pandas as pd
import numpy as np
```

```python
dataFile = "D:\\Git\\MachineLearning\\ScikitLearn\\Datasets\\ad-dataset\\ad.data"
data = pd.read_csv(dataFile,sep=",",header=None,low_memory=False)
```

```python
data.head(20)
```

```python
#check and mannipulate any missing value to NaN
def toNum(cell):
    try:
        return np.float(cell)
    except:
        return np.nan

#Apply missing values check to a column/Panda Series
def seriesToNum(series):
    return series.apply(toNum)
```

```python
train_data=data.iloc[0:,0:-1].apply(seriesToNum)
train_data.head(20)
```

```python
train_data=train_data.dropna()
```

```python
train_data.head(20)
```

```python
train_data.shape
```

```python
def toLabel(str):
    if str=="ad.":
        return 1
    else:
        return 0

train_label = data.iloc[train_data.index,-1].apply(toLabel)
train_label.head(20)   
        
```

```python
train_data.head(10)
```

```python
train_label.head(10)
```

```python
#Training Phase
from sklearn.svm import LinearSVC
classifier = LinearSVC()
classifier.fit(train_data[100:2300],train_label[100:2300])
```

```python
#Test Phase
# classifier.predict(train_data.iloc[12].reshape(1,-1))-- deprecated in pandas
classifier.predict(train_data.iloc[12].values.reshape(1,-1))
```
